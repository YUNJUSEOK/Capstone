```python
#selenium 설치
!pip install selenium

from selenium import webdriver
from selenium.webdriver.chrome.options import Options
import pandas as pd

#옵션 설정
options=Options()
options.add_argument('==kiosk')
options.add_argument('--start-fullscreen')
```

```python
#크롬 드라이버 불러오기
driver = webdriver.Chrome("chromedriver.exe")
```

```python
team_list=["NC","두산","KT","LG","키움","KIA","롯데","삼성","SK","한화","넥센"]
```

```python
data=pd.DataFrame()
```

```python
#스탯티즈 선수별 투구 기록 url 분할
url_1="http://www.statiz.co.kr/stat.php?opt=0&sopt=0&re=1&ys="
url_2="&ye="
url_3="&se=0&te="
url_4="&tm=&ty=0&qu=auto&po=0&as=&ae=&hi=&un=&pl=&da=1&o1=WAR&o2=OutCount&de=1&lr=0&tr=&cv=&ml=1&sn=30&si=&cn="
```

```python
# 각 팀마다 2001~2020년 타격데이터 수집
for team in team_list:
    for year in range(2001,2021):
        # url 접속
        driver.get(url_1+str(year)+url_2+str(year)+url_3+team+url_4)
        
        # 데이터 저장을 위한 빈 리스트 생성
        list_name=[]
        list_game=[]
        list_WAR=[]
        list_CG=[]
        list_SO=[]
        list_GS=[]
        list_W=[]
        list_L=[]
        list_SV=[]
        list_HLD=[]
        list_IP=[]
        list_R=[]
        list_ER=[]
        list_hitter=[]
        list_H=[]
        list_B2=[]
        list_B3=[]
        list_HR=[]
        list_BB=[]
        list_IBB=[]
        list_HBP=[]
        list_K=[]
        list_BK=[]
        list_WP=[]
        list_ERA=[]
        list_FIP=[]
        list_WHIP=[]
        list_ERA_=[]
        list_FIP_=[]
        
        #xpath를 통해 데이터 수집
        try:
            for i in range(3,32):
                if i==13 or i==14 or i==25 or i==26:
                    continue

                name=driver.find_element_by_xpath('//*[@id="fixcol"]/table/tbody/tr['+str(i)+']/td[2]/a')
                WAR=driver.find_element_by_xpath('//*[@id="fixcol"]/table/tbody/tr['+str(i)+']/td[4]/font/span')
                game=driver.find_element_by_xpath('//*[@id="mytable"]/tbody/tr['+str(i)+']/td[5]/a/font/span')
                CG=driver.find_element_by_xpath('//*[@id="mytable"]/tbody/tr['+str(i)+']/td[6]/a/font/span')
                SO=driver.find_element_by_xpath('//*[@id="mytable"]/tbody/tr['+str(i)+']/td[7]/a/font/span')
                GS=driver.find_element_by_xpath('//*[@id="mytable"]/tbody/tr['+str(i)+']/td[8]/a/font/span')
                W=driver.find_element_by_xpath('//*[@id="mytable"]/tbody/tr['+str(i)+']/td[9]/a/font/span')
                L=driver.find_element_by_xpath('//*[@id="mytable"]/tbody/tr['+str(i)+']/td[10]/a/font/span')
                SV=driver.find_element_by_xpath('//*[@id="mytable"]/tbody/tr['+str(i)+']/td[11]/a/font/span')
                HLD=driver.find_element_by_xpath('//*[@id="mytable"]/tbody/tr['+str(i)+']/td[12]/a/font/span')
                IP=driver.find_element_by_xpath('//*[@id="mytable"]/tbody/tr['+str(i)+']/td[13]/a/font/span')
                R=driver.find_element_by_xpath('//*[@id="mytable"]/tbody/tr['+str(i)+']/td[14]/font/span')
                ER=driver.find_element_by_xpath('//*[@id="mytable"]/tbody/tr['+str(i)+']/td[15]/font/span')
                hitter=driver.find_element_by_xpath('//*[@id="mytable"]/tbody/tr['+str(i)+']/td[16]/a/font/span')
                H=driver.find_element_by_xpath('//*[@id="mytable"]/tbody/tr['+str(i)+']/td[17]/a/font/span')
                B2=driver.find_element_by_xpath('//*[@id="mytable"]/tbody/tr['+str(i)+']/td[18]/a/font/span')
                B3=driver.find_element_by_xpath('//*[@id="mytable"]/tbody/tr['+str(i)+']/td[19]/a/font/span')
                HR=driver.find_element_by_xpath('//*[@id="mytable"]/tbody/tr['+str(i)+']/td[20]/a/font/span')
                BB=driver.find_element_by_xpath('//*[@id="mytable"]/tbody/tr['+str(i)+']/td[21]/a/font/span')
                IBB=driver.find_element_by_xpath('//*[@id="mytable"]/tbody/tr['+str(i)+']/td[22]/a/font/span')
                HBP=driver.find_element_by_xpath('//*[@id="mytable"]/tbody/tr['+str(i)+']/td[23]/a/font/span')
                K=driver.find_element_by_xpath('//*[@id="mytable"]/tbody/tr['+str(i)+']/td[24]/a/font/span')
                BK=driver.find_element_by_xpath('//*[@id="mytable"]/tbody/tr['+str(i)+']/td[25]/font/span')
                WP=driver.find_element_by_xpath('//*[@id="mytable"]/tbody/tr['+str(i)+']/td[26]/font/span')
                ERA=driver.find_element_by_xpath('//*[@id="mytable"]/tbody/tr['+str(i)+']/td[27]/a/font/span')
                FIP=driver.find_element_by_xpath('//*[@id="mytable"]/tbody/tr['+str(i)+']/td[28]/font/span')
                WHIP=driver.find_element_by_xpath('//*[@id="mytable"]/tbody/tr['+str(i)+']/td[29]/font/span')
                ERA_=driver.find_element_by_xpath('//*[@id="mytable"]/tbody/tr['+str(i)+']/td[30]/font/span')
                FIP_= driver.find_element_by_xpath('//*[@id="mytable"]/tbody/tr['+str(i)+']/td[31]/font/span')

                list_name.append(name.text)
                list_game.append(game.text)
                list_WAR.append(WAR.text)
                list_CG.append(CG.text)
                list_SO.append(SO.text)
                list_GS.append(GS.text)
                list_W.append(W.text)
                list_L.append(L.text)
                list_SV.append(SV.text)
                list_HLD.append(HLD.text)
                list_IP.append(IP.text)
                list_R.append(R.text)
                list_ER.append(ER.text)
                list_hitter.append(hitter.text)
                list_H.append(H.text)
                list_B2.append(B2.text)
                list_B3.append(B3.text)
                list_HR.append(HR.text)
                list_BB.append(BB.text)
                list_IBB.append(IBB.text)
                list_HBP.append(HBP.text)
                list_K.append(K.text)
                list_BK.append(BK.text)
                list_WP.append(WP.text)
                list_ERA.append(ERA.text)
                list_FIP.append(FIP.text)
                list_WHIP.append(WHIP.text)
                list_ERA_.append(ERA_.text)
                list_FIP_.append(FIP_.text)
        
        #데이터가 없는경우 다음 루프 진행
        except:
            continue
        
        #리스트를 데이터프레임으로 변환 후 저장
        player_data=pd.DataFrame({'이름':list_name,'출장':list_game,
                                '완투':list_CG,'완봉':list_SO,
                                '선발':list_GS,'승':list_W,
                                 '패':list_L,'세':list_SV,
                                '홀드':list_HLD,'이닝':list_IP,
                                '실점':list_R,'자책':list_ER,
                                '타자':list_hitter,'안타':list_H,
                                '2타':list_B2,'3타':list_B3,
                                '홈런':list_HR,'볼넷':list_BB,
                                '고4':list_IBB,'사구':list_HBP
                                ,'삼진':list_K,'보크':list_BK,
                                '폭투':list_WP,'ERA':list_ERA,
                                'FIP':list_FIP,'WHIP':list_WHIP,
                                'ERA+':list_ERA_,'FIP+':list_FIP_,
                                 'WAR':list_WAR})
        data=data.append(player_data)
```

```python
#csv파일로 변환
data.to_csv("baseball_pitcher_data.csv",encoding="euc-kr",index=None)
```
